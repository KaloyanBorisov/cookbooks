```python
# Example customer query that spans multiple BUs
customer_query = """
Hi, I ordered a laptop 3 days ago (Order ID: ORD-12345) but it hasn't shipped yet. 
My card shows a $1,299 charge but the website said $1,199. 
Also, I forgot to apply my $50 promo code 'FALL50' - can you still apply it?
"""

# Invoke the main supervisor graph
result = await graph.ainvoke({
    "messages": [HumanMessage(content=customer_query)]
})
for msg in result["messages"]:
    msg.pretty_print()
```
